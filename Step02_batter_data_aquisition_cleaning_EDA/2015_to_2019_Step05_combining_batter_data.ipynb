```python
# Create dictionary with key as existing column and the value as the new, cumsum column
# Will use to loop creation of new features

cs = ['total_pitches', 'hits', 'abs', 'whiffs', 'swings', 'takes', 'k', 'walk', 'single', 'double',
              'triple', 'hr', 'line_drive', 'ground_ball', 'fly_ball', 'popup', 'rbi', 'sac']

cs_dict = {}

for c in cs:
    new_col = c + '_cs'
    cs_dict.update({c: new_col})
```

```python
for c in cs_dict:
    batters[cs_dict[c]] = batters.groupby(['player_name', batters.game_date.dt.year])[c].cumsum()
```

## Calculate standard basball performance metrics back into the dataframe
These include:
- batting average (ba)
- slugging percentage (slg)
- isolated power (iso)
- batting average on balls in play (babip)

### Create functions for calculations

```python
def batting_average(hits, at_bats):
    if at_bats == 0:
        return 0
    else:
        return hits / at_bats

def slugging_percentage(single, double, triple, home_run):
    return single + (2 * double) + (3 * triple) + (4 * home_run)

def isolated_power(slg, ba):
    return slg - ba

def ba_balls_in_play(hits, home_run, at_bat, strike_out, sacrifice):
    return (hit - home_run) / (at_bat - home_run - strike_out + sacrifice)
```

```python
batters.columns
```

```python
batters['ba_cs'] = batters.hits_cs / batters.abs_cs
```

```python
batters['slg_cs'] = (batters.single_cs \
                     + 2 * batters.double_cs \
                     + 3 * batters.triple_cs \
                     + 4 * batters.hr_cs)/ \
                        batters.abs_cs
```

```python
batters['iso_cs'] = batters.slg_cs - batters.ba_cs
```

```python
batters['babip_cs'] = (batters.hits_cs - batters.hr_cs) / \
                        (batters.abs_cs - batters.hr_cs - batters.k_cs + batters.sac_cs)
```

## Chop dataframe into each year

```python
batters.game_date = batters.game_date.apply(lambda d: pd.to_datetime(d))
```

```python
b_2019 = batters.loc[batters.game_date.dt.year == 2019]
b_2018 = batters.loc[batters.game_date.dt.year == 2018]
b_2017 = batters.loc[batters.game_date.dt.year == 2017]
b_2016 = batters.loc[batters.game_date.dt.year == 2016]
b_2015 = batters.loc[batters.game_date.dt.year == 2015]
```

```python
b_year = {'y2019': ['2019-01-01', b_2019], 
          'y2018': ['2018-01-01', b_2018],
          'y2017': ['2017-01-01', b_2017], 
          'y2016': ['2016-01-01', b_2016], 
          'y2015': ['2015-01-01', b_2015]}
```

```python
DF = batters.groupby(by = ['player_id', 'player_name'], axis = 0).sum()
DF = DF.reset_index()
```

```python
dummy_df_list = []

for y in b_year.items():
    date = y[1][0]
    bats = y[1][1]
    
    dummy_df = pd.DataFrame(columns = bats.columns)
    dummy_df.player_id = DF.player_id
    dummy_df.player_name = DF.player_name
    dummy_df.game_date = date
    dummy_df.game_date = pd.to_datetime(dummy_df.game_date)
    dummy_df = dummy_df.fillna(value = 0)
    
    batters_dummy = pd.concat([bats, dummy_df], axis = 0)
    batters_dummy = batters_dummy.sort_values(by=['player_id', 'game_date'])
    batters_dummy = batters_dummy.reset_index()
    batters_dummy['shift_date'] = batters_dummy.game_date.shift(-1)

    batters_dummy = batters_dummy[['player_id', 'game_date', 'shift_date', 'player_name', 'total_pitches', 
                   'hits', 'abs', 'whiffs', 'swings', 'takes', 'k', 'walk', 'single', 
                   'double', 'triple','hr', 'line_drive', 'ground_ball', 'fly_ball', 'popup', 
                   'rbi', 'sac','ba', 'slg', 'iso', 'babip']]
    
    batters_dummy = batters_dummy.loc[batters_dummy.shift_date != date][0:-1]
    
    dummy_df_list.append(batters_dummy)
```
