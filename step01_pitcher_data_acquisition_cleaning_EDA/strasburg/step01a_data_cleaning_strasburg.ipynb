#### These velocities relate directly to pitch types, let's take a look at these values together

```python
pitches = stras.pitch_type.unique()
pitches
```

```python
# Where are these null values?
stras.loc[stras.pitch_type.isnull() == True]
```

```python
# Only one pitch with no pitch_type.  We will drop this one.
stras = stras.loc[stras.pitch_type.isnull() == False]

pitches = stras.pitch_type.unique()
pitches
```

```python
results = {}
p_count = 1
for p in pitches:
    avg = stras.loc[(stras.pitch_type == p)].release_speed.mean()
    low = stras.loc[(stras.pitch_type == p)].release_speed.min()
    high = stras.loc[(stras.pitch_type == p)].release_speed.max()
    i = str('p{}'.format(p_count))
    results[i] = [p, avg, low, high]
    p_count += 1 
```

```python
results = pd.DataFrame(results).T
results.columns = ['pitch_type', 'mean', 'min', 'max']
results
```
