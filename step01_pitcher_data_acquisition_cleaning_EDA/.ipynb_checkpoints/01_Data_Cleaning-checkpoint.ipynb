```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import pickle
%matplotlib inline
```

```python
filepath = '../data/savant/'

infile = open(filepath + 'all_pitcher_data_2019.pickle','rb')
pitchers = pickle.load(infile)
infile.close()
```

```python
pitchers.head();
```

```python
pitchers.columns
```

```python
keep_cols = ['pitch_type', 'sv_id', 'game_date', 'player_name', 'batter', 'pitcher','stand', 'p_throws', 
             'balls', 'strikes', 'game_year', 'on_3b', 'on_2b', 'on_1b', 'outs_when_up', 'inning', 
             'inning_topbot', 'pitch_number', 'home_score', 'away_score']

kill_cols = []

for k in pitchers.columns:
    if k in keep_cols:
        pass
    else:
        kill_cols.append(k)
```

```python
pitchers = pitchers.drop(columns = kill_cols)
```

```python
pitchers
```

## Fill nulls

```python
pitchers.isnull().sum()
```

```python
pitchers = pitchers.dropna(subset=['pitch_type'])
pitchers = pitchers.dropna(subset=['sv_id'])
```

```python
pitchers.on_3b = pitchers.on_3b.fillna(value=0)
pitchers.on_2b = pitchers.on_2b.fillna(value=0)
pitchers.on_1b = pitchers.on_1b.fillna(value=0)
```

```python
pitchers.isnull().sum()
```

```python
pitchers
```

```python
pitchers.game_date = pd.to_datetime(pitchers.game_date)
```

```python
pitchers['stand_0right'] = pitchers.stand.apply(lambda c: 0 if c == 'R' else '1')
pitchers = pitchers.drop(columns = 'stand')

pitchers.on_3b = pitchers.on_3b.apply(lambda c: 0 if c == 0 else 1)
pitchers.on_2b = pitchers.on_2b.apply(lambda c: 0 if c == 0 else 1)
pitchers.on_1b = pitchers.on_1b.apply(lambda c: 0 if c == 0 else 1)

pitchers['pitcher_home0'] = pitchers.inning_topbot.apply(lambda c: 0 if c == 'Top' else 1)
pitchers = pitchers.drop(columns = 'inning_topbot')

pitchers['p_throws_0right'] = pitchers.p_throws.apply(lambda c: 0 if c == 'R' else 1)
pitchers = pitchers.drop(columns = 'p_throws')
```

```python
pitchers
```

## Feature Engineering

```python
# Sort table cronologically by pitcher
pitchers = pitchers.sort_values(by = ['sv_id'])
```

### Create running total of pitches thrown in the season per pitcher

```python
pitchers['pitch'] = 1
```

```python
pitchers['pitch_season'] = pitchers.groupby(['player_name'])['pitch'].cumsum() - 1
```

### Total pitches per game

```python
pitchers['pitch_game'] = pitchers.groupby(['player_name', 'game_date'])['pitch'].cumsum() - 1
```

### Create running total of pitches thrown to each batter per game

```python
pitchers['pitch_bat_gm'] = pitchers.groupby(['player_name', 'game_date', 'batter'])['pitch'].cumsum() - 1
```

### Pitches per current at bat

```python
pitchers['pitch_ab'] = pitchers.pitch_number - 1
pitchers = pitchers.drop(columns = ['pitch_number', 'pitch', 'sv_id'])
```

```python
pitchers.loc[pitchers.player_name == 'Justin Verlander']
```

```python
# Reorder columns
pitchers = pitchers[['pitch_type', 'game_date', 'game_year', 'player_name', 'batter', 'pitcher', 
                     'balls', 'strikes', 'outs_when_up', 'on_3b', 'on_2b', 'on_1b', 'inning', 
                     'home_score', 'away_score', 'stand_0right', 'pitcher_home0', 'p_throws_0right', 
                     'pitch_season', 'pitch_game', 'pitch_bat_gm', 'pitch_ab']]
```

```python
filepath = '../data/processed/pitchers/'

outfile = open(filepath + 'cleaned.pickle','wb')
pitchers = pickle.dump(pitchers, outfile)
outfile.close()
```
