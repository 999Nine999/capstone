```python
import pandas as pd
import numpy as np
import pickle
```

```python
infile1 = open('../data/processed/batters/2019_batters_date_shifted.pickle','rb')
batters = pickle.load(infile1)
infile1.close()
```

```python
infile2 = open('../data/processed/scherzer/scherzer_clean_new_features_2019.pickle','rb')
pitcher = pickle.load(infile2)
infile2.close()
```

```python
batters = batters.rename(columns = {'player_id': 'batter_id'})
```

```python
pitcher = pitcher.rename(columns = {'batter': 'batter_id'})
```

```python
pitcher.game_date = pd.to_datetime(pitcher.game_date)
batters.shift_date = pd.to_datetime(batters.shift_date)
```

```python
pb = pitcher.merge(batters, 
                   how = 'left', 
                   left_on = ['game_date', 'batter_id'], 
                   right_on = ['shift_date', 'batter_id'])
```

```python
outfile = open('../data/processed/scherzer/scherzer_with_batters_2019.pickle','wb')
pickle.dump(pb ,outfile)
outfile.close()
```
